#### 4. Проверьте гипотезу о том, что ожидаемые результаты экзамена по теории вероятностей тех, у кого фамилия начинается с гласной буквы и с согласной буквы, равны. В качестве альтернативной гипотезы возьмите гипотезу о неравенстве.

```python
df = pd.read_csv('22-23_hse_probability - main.csv')
display(df)
```

Гипотезы:

$$
\begin{aligned}
& H_0: \mu_{гласные} = \mu_{согласные} \\
& H_1: \mu_{гласные} \neq \mu_{согласные}
\end{aligned}
$$

Уровень значимости: 5%

```python
gl = ['А', 'Е', 'Ё', 'И', 'О', 'У', 'Ы', 'Э', 'Ю', 'Я']
glas = df.loc[df["Last name"].str.startswith(tuple(gl), na=False)]["Экзамен"]
glas_df = pd.DataFrame(glas.dropna())
glas_df = glas_df.replace("неявка", 0)
glas_df = glas_df.replace(",", ".", regex=True)
glas_df = glas_df.astype(float)
#display(glas_df)
#glas.shape()

sogl = ['Б', 'В', 'Г', 'Д', 'Ж', 'З', 'Й', 'К', 'Л', 'М', 'Н', 'П', 'Р', 'С', 'Т', 'Ф', 'Х', 'Ц', 'Ч', 'Ш', 'Щ']
soglas = df[df["Last name"].str.startswith(tuple(sogl), na=False)]["Экзамен"]
soglas_df = pd.DataFrame(soglas.dropna())
soglas_df = soglas_df.replace("неявка", 0)
soglas_df = soglas_df.replace(",", ".", regex=True)
soglas_df = soglas_df.astype(float)
#display(soglas_df)

```

а) [5] Используйте тест Уэлча.

```python
from scipy.stats import ttest_ind

print(ttest_ind(glas_df, soglas_df))

pvalue1 = ttest_ind(glas_df, soglas_df) [1]
print("p-value:", pvalue1)

if pvalue1 < 0.05:
    print("H0 отвергается")
else:
    print("Нет оснований отвергать H0")
    
```

Таким образом, можно утверджать, что то ожидаемые результаты экзамена по теории вероятностей тех, у кого фамилия начинается с гласной буквы и с согласной буквы, равны. 

#### 5. Составьте таблицу сопряжённости, поделив студентов писавших экзамен на четыре группы по двум признакам: набрал ли больше медианы или нет, на согласную или гласную букву начинается фамилия.

```python
df_ekz = df["Экзамен"].dropna().replace("неявка", 0)
df_ekz = df_ekz.str.replace(",", ".").astype(float)
med = df_ekz.median()

print(med)
```

```python
bm_g = (glas_df >= med).sum()
bm_s = (soglas_df >= med).sum()
mm_g = (glas_df < med).sum()
mm_s = (soglas_df < med).sum()

table = pd.DataFrame({
    ">= медианы": [bm_g.values[0], bm_s.values[0]],
    "< медианы": [mm_g.values[0], mm_s.values[0]] 
})

table.index = ["гласная", "согласная"]

display(table)
```

а) [5] Постройте 95% асимптотический интервал для отношения шансов хорошо написать экзамен («несогласных» к «согласным»). Проверьте гипотезу о том, что отношение шансов равно 1 и укажите P-значение.

```python
from scipy.stats import norm
```

```python
o_r =(bm_g[0]/(mm_g[0]) ) / (bm_s[0]/(mm_s[0]))

print("Отношения шансов хорошо написать экзамен (гласных к согласным):", o_r)
```

```python
se = np.sqrt((1 / mm_s.values[0]) + (1 / mm_g.values[0]) + (1 / bm_s.values[0]) + (1 / bm_g.values[0]))

low = np.exp(np.log(o_r) - 1.96 * se)
up = np.exp(np.log(o_r) + 1.96 * se)
print("95%-ый доверительный интервал:", low, up)
```

б) [5] Постройте 95% асимптотический интервал для отношения вероятностей хорошо написать экзамен. Проверьте гипотезу о том, что отношение вероятностей равно 1 и укажите P-значение.

```python
p =(bm_g[0]/(bm_g[0]+ mm_g[0]) ) / (bm_s[0]/(bm_s[0]+mm_s[0]))

print("Отношения вероятностей хорошо написать экзамен (гласных к согласным):", p)
```

```python
se2 = math.sqrt((1 / (bm_g[0]+ mm_g[0])) + (1 / (bm_s[0]+mm_s[0])))

low2 = np.exp(np.log(p) - 1.96 * se2)
up2 = np.exp(np.log(p) + 1.96 * se2)
print("95%-ый доверительный интервал:", low2, up2)
```

```python
z2 = np.log(p) / se2
p_value = 2 * (1 - norm.cdf(abs(z2)))
print("p-value:", p_value)

if p_value < 0.05:
    print("H0 отвергается")
else:
    print("Нет оснований отвергать H0")
```
